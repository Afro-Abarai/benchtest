```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
```

Now data is loaded from train.csv to a pandas dataframe

```python
df = pd.read_csv('train.csv')
#shape of data
df.shape

```

NOw we have y as output column. Next target/output column is extracted from dataframe in a separate array and then removed from dataset

```python
y = df['y'].values
y
```

Now a/c to problem statement all the features required for benchtest starts with X. So now we extract all the features necessary.


```python
cols = [x for x in df.columns if 'X' in x]
print('Necessary features: {}'.format(len(cols)))
cols
```

Now we need to find out whether there is any feature which doesn't have numerical values & if that exists then we need to change those values into numerical ones.

```python
df[cols].dtypes.value_counts()
```

Now data in each of the column is counted

```python
data_count_cols = [[], [], []]
for c in cols:
    typ = df[c].dtype
    uniq = len(np.unique(df[c]))
    if uniq == 1:
        data_count_cols[0].append(c)
    elif uniq == 2 and typ == np.int64:
        data_count_cols[1].append(c)
    else:
        data_count_cols[2].append(c)
```

```python
print('Constant features: {} \nBinary features: {} \nCategorical features: {}\n'
      .format(*[len(c) for c in data_count_cols]))
print('Constant features:', data_count_cols[0])
print('Categorical features:', data_count_cols[2])
```

Now I load test data in a pandas dataframe


```python
df_test = pd.read_csv('test.csv')

# remove\ing columns ID and output data from the dataframe as they are not required in model training
usable_columns = list(set(df.columns) - set(['ID', 'y']))
y_train = df['y'].values
id_test = df_test['ID'].values

x_train = df[usable_columns]
x_test = df_test[usable_columns]
```

Now I perform data cleaning by running a check for null and unique values on datasets.

```python
def check_missing_values(df):
    if df.isnull().any().any():
        print("There are missing values in the dataframe")
    else:
        print("There are no missing values in the dataframe")
check_missing_values(x_train)
check_missing_values(x_test)
```

Now I apply label encoding on columns and remove columns with variance equal to zero.

```python
for column in usable_columns:
    cardinality = len(np.unique(x_train[column]))
    if cardinality == 1:
        x_train.drop(column, axis=1) # all the columns with only one value is dropped
        x_test.drop(column, axis=1)
    if cardinality > 2: # categorical column check
        mapper = lambda x: sum([ord(digit) for digit in x])
        x_train[column] = x_train[column].apply(mapper)
        x_test[column] = x_test[column].apply(mapper)
x_train.head()
```

```python
# checking the data type for numerical values

print('Feature types:')
x_train[cols].dtypes.value_counts()
```

applied PCA for linear dimensionality reduction

```python
n_comp = 12
pca = PCA(n_components=n_comp, random_state=420)
pca2_results_train = pca.fit_transform(x_train)
pca2_results_test = pca.transform(x_test)
```

Initialising xgboost

```python
import xgboost as xgb
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split

x_train, x_valid, y_train, y_valid = train_test_split(
        pca2_results_train, 
        y_train, test_size=0.2, 
        random_state=4242)

d_train = xgb.DMatrix(x_train, label=y_train)
d_valid = xgb.DMatrix(x_valid, label=y_valid)

d_test = xgb.DMatrix(pca2_results_test)
```

```python
params = {}
params['objective'] = 'reg:linear'
params['eta'] = 0.02
params['max_depth'] = 4

def xgb_r2_score(preds, dtrain):
    labels = dtrain.get_label()
    return 'r2', r2_score(labels, preds)

watchlist = [(d_train, 'train'), (d_valid, 'valid')]

clf = xgb.train(params, d_train, 
                1000, watchlist, early_stopping_rounds=50, 
                feval=xgb_r2_score, maximize=True, verbose_eval=10)
             
```

```python
#finally predicting the test dataframe values using xgboost

p_test = clf.predict(d_test)
p_test
```
